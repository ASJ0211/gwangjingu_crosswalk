```python
from math import radians, sin, cos, sqrt, atan2

# haversine 함수 정의
def haversine(lat1, lon1, lat2, lon2):
    R = 6371.0 # 지구의 반지름 (km)
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c * 1000 # 두 지점간의 거리 (meter)
    return distance

# 거리 계산 및 distance 컬럼 추가
df['distance'] = df.apply(lambda x: haversine(x['lat1'], x['lon1'], x['lat2'], x['lon2']), axis=1)
df = df.sort_values(by='링크 길이', ascending=False)
# 결과 확인
print(df.head())

```
